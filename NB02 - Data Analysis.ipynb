```python
import pandas as pd
import yfinance as yf
from fredapi import fred


```

```python
df = pd.read_csv("./data/Data.csv", index_col=0, parse_dates=True)
tickers = list(df.columns)
```

```python
sector_data = []

for ticker in tickers:
    try:
        info = yf.Ticker(ticker).info
        sector = info.get("sector", "Unknown")
        sector_data.append({"Ticker": ticker, "Sector": sector})
    except Exception as e:
        print(f"Error for {ticker}: {e}")
        sector_data.append({"Ticker": ticker, "Sector": "Unknown"})

# Assign sector to each ticker
sector_df = pd.DataFrame(sector_data).set_index("Ticker")

# One-hot encode
sector_ohe = pd.get_dummies(sector_df["Sector"]).astype(int)
```
